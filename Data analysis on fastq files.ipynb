```python
def Qtophred33(q):
    v=33+q
    ask=chr(v)
    return ask
```

```python
def phred33toQ(ask):
    v=ord(ask)
    q=v-33  
    return q
```

```python
j=list("")
```

```python
for i in qual:
    for char in i:
        j.append(phred33toQ(char))
```
